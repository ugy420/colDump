### 3. Plot the histogram of the Age of the passengers by their sex:

(i) Stacked histogram \
(ii) Side by side histogram

```python
# (i)
sa=df.pivot(columns='sex').age
plt.hist(sa,16, stacked=True, label = ['Female', 'Male'], color=['red', 'blue'])
plt.legend(loc='right')
plt.title('Stacked Histogram')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# (ii)
sa=df.pivot(columns='sex').age
plt.hist(sa,16, stacked=False, label = ['Female', 'Male'], color=['red', 'blue'])
plt.legend(loc='right')
plt.title('Side by Side Histogram')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

### 6. Show the Box and Whisker plot of the age of the passengers, side by side based on the condition, whether:

(iii) the passenger was male or female
